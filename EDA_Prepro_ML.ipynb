```python
######################## General function to detect/fill in the NaN values with the mode ###################
```

```python
def fill_na(data, column_name):
    import pandas as pd
    mode_value = data[column_name].mode()[0]
    data[column_name].fillna(mode_value, inplace=True)
    return data
```

```python
def drop_na(data, column_name):
    import pandas as pd
    data.drop(column_name, axis=1, inplace=True)
    return data
```

```python
def get_dummies(data, column_name): 
    
    import pandas as pd
    
    dummies = pd.get_dummies(data[column_name],
                             prefix=column_name,
                             drop_first=True)
    
    
    data = pd.concat([data, dummies],
                     axis=1)
    
    # Löschen der ursprünglichen Spalte
    data.drop(column_name,
              axis=1,
              inplace=True)
    
    return data
```

```python
def fill_na_with_mean(data, column_name):
    mean_value = data[column_name].mean()
    data[column_name].fillna(mean_value, inplace=True)
    return data
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_data = df[categorical_cols]

# OneHotEncoder instanziieren
encoder = OneHotEncoder(drop='first', sparse=False)  # drop='first' verhindert die Multi-Kollinearität

# Encoder an die kategorialen Daten anpassen und transformieren
encoded_data = encoder.fit_transform(categorical_data)

# Ergebnis in einen DataFrame umwandeln
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Falls Sie den encodierten DataFrame mit dem originalen DataFrame kombinieren möchten
df = df.drop(columns=categorical_cols)
df = pd.concat([df, encoded_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df.columns
```

```python
df.info()
```

```python
y = df.Transported.copy()
train_data = df.drop(['Transported'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(train_data, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
svm_pred = svc.predict(X_test)
svc_accuracy = accuracy_score(svm_pred, y_test)

svc_accuracy
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()
sgd.fit(X_train, y_train)
sgd_pred = sgd.predict(X_test)
sgd_accuracy = accuracy_score(sgd_pred, y_test)

sgd_accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
decision_tree_pred = decision_tree.predict(X_test)
decision_tree_accuracy = accuracy_score(decision_tree_pred, y_test)

decision_tree_accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
random_forest_pred = random_forest.predict(X_test)
random_forest_accuracy = accuracy_score(random_forest_pred, y_test)

random_forest_accuracy
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(random_state = 42)
gbc.fit(X_train, y_train)
gbc_pred = gbc.predict(X_test)
gbc_accuracy = accuracy_score(gbc_pred, y_test)

gbc_accuracy
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'Stochastic Gradient Decent', 'Decision Tree','Random Forest',
              'GradientBoostingClassifier'],
    
    'Accuracy': [svc_accuracy, sgd_accuracy, decision_tree_accuracy, random_forest_accuracy, gbc_accuracy]})

models.sort_values(by='Accuracy', ascending=False)
```
